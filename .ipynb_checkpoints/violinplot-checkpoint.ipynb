```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', scale='count')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', scale='count', scale_hue=False)
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', scale='count', dodge=False)
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', scale='count', split=True, scale_hue=False)
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', inner='box')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', inner='quartile')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', inner='point')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', inner='stick')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', inner=None)
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', bw=0.3)
```
