```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights = flights.pivot('month', 'year', 'passengers')
flights
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(flights)
plt.xticks(rotation=0)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights, col_cluster=False)
```

```python
sns.clustermap(flights, row_cluster=False)
```

```python
sns.clustermap(flights, cmap='coolwarm', linewidth=0.01)
```

```python
sns.clustermap(flights, cmap='Reds', linewidth=0.01)
```
