```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
cov = [[0.2, 0], [0, 3]]
```

```python
x_axis, y_axis = np.random.multivariate_normal(mean, cov, size=40).T
```

```python
sns.kdeplot(x_axis)
```

```python
sns.kdeplot(y_axis)
```

```python
sns.kdeplot(x_axis, label='x')
sns.kdeplot(y_axis, label='y')
```

```python
sns.kdeplot(x_axis, shade=True, color='purple')
```

```python
sns.kdeplot(x_axis, y_axis, shade=True)
```

```python
sns.kdeplot(x_axis, y_axis, cmap='RdBu_r')
```

```python
sns.kdeplot(x_axis, y_axis, cmap='coolwarm', n_levels=20)
```

```python
sns.kdeplot(x_axis, bw=0.1)
```

```python
sns.kdeplot(x_axis, vertical=True)
```
