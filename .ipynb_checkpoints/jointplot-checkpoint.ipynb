```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
iris.head()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='reg')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg', color='g')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='resid')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='kde', color='r')
```

```python
from scipy.stats import spearmanr, pearsonr
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', stat_func=spearmanr)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', stat_func=pearsonr)
```

```python
sns.jointplot(data=tips, x='size', y='total_bill', stat_func=spearmanr)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', ratio=1)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', ratio=10)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', ratio=1, size=10)
```
