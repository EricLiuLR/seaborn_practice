```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
g = sns.FacetGrid(tips, row='smoker', col='time')
```

```python
g = sns.FacetGrid(tips, row='smoker', col='time')
g.map(plt.hist, 'total_bill')
```

```python
g = sns.FacetGrid(tips, row='smoker', col='sex')
g.map(sns.violinplot, 'total_bill')
```

```python
g = sns.FacetGrid(tips, row='smoker', col='sex')
g.map(sns.scatterplot, 'tip', 'total_bill', marker='D')
```

```python
g = sns.FacetGrid(tips, row='smoker', col='sex')
g.map(sns.violinplot, 'time', 'total_bill', scale='count')
```

```python
g = sns.FacetGrid(tips, row='sex', col='smoker')
g.map(sns.regplot, 'total_bill', 'tip')
```

```python
g = sns.FacetGrid(tips, col='time', hue='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col='day', size=2)
g.map(sns.boxplot, 'time', 'total_bill')
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col='day', size=7, aspect=0.4)
g.map(sns.boxplot, 'time', 'total_bill', 'sex')
g.add_legend()
```
