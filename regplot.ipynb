```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', logx=True)
```

```python
mean = [2, 5]
cov = [[1.1, 0.4], [2.2, 3]]

x_value, y_value = np.random.multivariate_normal(mean, cov, 100).T
```

```python
sns.regplot(x=x_value, y=y_value, marker='D', scatter_kws={'s': 8})
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', marker='*',\
            scatter_kws={'s': 8, 'color': '0.7'}, line_kws={'color': 'r', 'linewidth': 1})
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', ci=34)
```

```python
sns.regplot(data=tips, x='size', y='total_bill')
```

```python
sns.regplot(data=tips, x='size', y='total_bill', x_jitter=0.1, scatter_kws={'s': 5})
```

```python
sns.regplot(data=tips, x='size', y='total_bill', x_estimator=np.mean)
```
