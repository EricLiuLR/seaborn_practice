```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
g = sns.PairGrid(data=iris)
g.map(plt.scatter, **{'s': 2})
```

```python
g = sns.PairGrid(iris, hue='species')
g.map_diag(plt.hist, **{'bins':20})
g.map_offdiag(plt.scatter, **{'s': 5})
g.add_legend()
```

```python
g = sns.PairGrid(iris, hue='species')
g.map_diag(sns.distplot, hist=False)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(plt.hist, histtype='step', bins=20, linewidth=2)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, vars=['petal_length', 'petal_width'], hue='species')
g.map_diag(plt.hist, histtype='step', linewidth=2)
g.map_offdiag(sns.scatterplot, alpha=0.8, edgecolor='black')
```

```python
g = sns.PairGrid(data=iris, x_vars=['petal_length', 'petal_width'], y_vars=['sepal_length', 'sepal_width'], hue='species')
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(plt.hist, histtype='step')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(data=iris, hue='species', hue_kws={'marker': ['d', 's', '*']})
g.map_diag(plt.hist, histtype='step')
g.map_offdiag(sns.scatterplot)
g.add_legend()
```
