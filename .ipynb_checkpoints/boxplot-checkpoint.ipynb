```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.boxplot(data=tips, x='size')
```

```python
sns.boxplot(data=tips, x='sex', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', palette='muted')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='time', palette='husl')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='time', order=['Thur', 'Sun', 'Sat', 'Fri'], hue_order=['Dinner', 'Lunch'], palette='husl')
```
