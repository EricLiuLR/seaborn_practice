```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
num = np.random.rand(12, 15)
sns.heatmap(num)
```

```python
sns.heatmap(num, annot=True)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(num, annot=True, linewidth=0.01)
```

```python
sns.heatmap(num, vmin=0, vmax=2, cmap='coolwarm')
```

```python
sns.heatmap(num, vmin=0, vmax=0.6, cmap='coolwarm')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights = flights.pivot(index='month', columns='year', values='passengers')
```

```python
flights
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, linewidth=0.01, annot=True, fmt='d')
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, linewidths=0.01, annot=True, fmt='d', vmin=100, vmax=650)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, cmap='RdBu', linewidth=0.01, annot=True, fmt='d')
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, cmap='Greens', linewidth=0.01, annot=True, fmt='d')
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, linewidth=0.01, annot=True, fmt='d', center=112, cmap='RdBu')
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(flights, linewidth=0.01, annot=True, fmt='d', cmap='RdBu', cbar=False)
```
