```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris, hue='species')
```
